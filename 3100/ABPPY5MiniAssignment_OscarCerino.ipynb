### 1. Suppose we have 3 datasets: IceCreamData.csv (from last class Mini Assignment) and 2 new files IceCreamDataNew.csv and CustomerData.csv

```python
#First, load the IceCreamData.csv file into a DataFrame and check the result with the .head() and .tail()
import pandas as pd

df_IC = pd.read_csv("./IceCreamData.csv")
print(df_IC.head())
print(df_IC.tail())
```

```python
#Second, join the New ice cream data so that it is at the bottom of the original Ice Cream Data.
# Read in the new data and add those rows your original DataFrame. 
# How many rows are in the new table? Check the result with the .tail() and also find the len() of the new DataFrame
df_ICNew = pd.read_csv('./IceCreamDataNew.csv')
df_IceCream = pd.concat([df_IC,df_ICNew])
print(df_IceCream.tail())
print(len(df_IceCream))
```

```python
#Finally, merge the data using the Customer ID with the last file: CustomerData.csv
df_CD = pd.read_csv("./CustomerData.csv")
df_IceCream = pd.merge(df_CD, df_IceCream, on='Customer')
df_IceCream
```

### 2. Now that you have the final version of the Ice Cream data, find the following summaries

```python
#Remove the Customer column
df_IceCream = df_IceCream.drop("Customer", axis=1)
df_IceCream
```

```python
# Create a summary grouping by Flavor showing means of Age
groupedd1 = df_IceCream["Age"].groupby(df_IceCream['Flavor'])
groupedd1.mean()
```

```python
# Create a summary grouping by Store showing means of Sales
groupedd2 = df_IceCream["Sales"].groupby(df_IceCream['Store'])
groupedd2.mean()
```

```python
# Create a summary grouping by Gender showing means of Age
groupedd3 = df_IceCream["Age"].groupby(df_IceCream["Gender"])
groupedd3.mean()
```

```python
# Create a summary grouping by Gender showing medians of Age
groupedd4 = df_IceCream["Age"].groupby(df_IceCream['Gender'])
groupedd4.median()
```

```python
# Create a summary grouping by Flavor showing means of Age and Sales (the only quant variables)
AgeSales = df_IceCream.groupby(['Flavor'])[["Age","Sales"]].mean()
AgeSales
```

```python
# Create the above summary so that the means have the prefix 'mean_'
AgeSales=AgeSales.add_prefix('mean_')
AgeSales
```

```python
# Create a new summary that groups by Flavor and then type and shows the mean of Sales
groupedd5 = df_IceCream.groupby([df_IceCream["Flavor"], df_IceCream["Type"]])
groupedd5.Sales.mean()
```

```python
# Create a summary that groups by Flavor and returns the Age and Sales mean and count
groupedd6 = df_IceCream.groupby(df_IceCream["Flavor"])
print(groupedd6.Age.count())
print(groupedd6.Age.mean())
print(groupedd6.Sales.count())
print(groupedd6.Sales.mean())
```

```python
#Create a summary that groups by Flavor and returns the mean and standard deviation of sales.
# But make sure to use the friendly titles "Average" and "Standard Deviation"
groupedd7 = df_IceCream.groupby(df_IceCream["Flavor"])
print(groupedd7.Sales.mean())
print(groupedd7.Sales.std())
```

```python
#Finally create a pivot table that returns the counts with Store as the rows
# and Type as the columns. Include the margin totals
pivot_IC = df_IceCream.pivot_table(index='Store', columns="Type", margins=True)
pivot_IC
```
